Import `pandas` as `pd` and `numpy` as `np`. Download the dataset `EPC_Industry_Analytics_Demo.xlsx` from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing. Load the dataset `EPC_Industry_Analytics_Demo.xlsx` into a DataFrame named `df`.

### Question 3: Missing Value Treatment (2 Marks)

First, replace any occurrences of blank spaces `' '`, empty strings `''`, `'None'`, and `'-'` with `np.nan` across the entire DataFrame.
Next, fill the missing values in all **numeric columns** with the median of their respective columns.
Finally, print the first 5 rows of the DataFrame to verify the changes.

**Expected Output:** A table showing the first 5 rows of the modified DataFrame.

### Question 6: Outlier Detection using IQR (3 Marks)

Write a loop to calculate the Interquartile Range (IQR) for each numeric column in `df`.
Define outliers as values below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR`.
Count the number of outliers for each numeric column and store them in a dictionary named `outlier_summary`. Print this dictionary.

**Expected Output:** A dictionary showing the column names as keys and their respective outlier counts as values.

```python
# Initialize the dictionary to store counts
outlier_summary = {}

# Select only the numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Loop through each numeric column
for col in numeric_cols:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers (values outside the bounds)
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_summary[col] = len(outliers)

# Print the final dictionary
print(outlier_summary)
```

### Question 11: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
from sklearn.datasets import load_digits

# Load the MNIST-style 'digits' dataset from sklearn
digits = load_digits()

# Separate into features (X) and target labels (y)
X = digits.data
y = digits.target

# Print the shapes
print(f"Shape of features (X): {X.shape}")
print(f"Shape of target (y): {y.shape}")
```

### Question 12: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans

# 1. Initialize the K-Means model with 10 clusters (for digits 0-9)
# n_init='auto' is the modern standard for cleaner execution
kmeans = KMeans(n_clusters=10, n_init='auto', random_state=42)

# 2. Fit the model on your features (X)
kmeans.fit(X)

# 3. Assign the resulting cluster labels
kmeans_labels = kmeans.labels_

# Verification: Print the first 20 labels to see the result
print("First 20 cluster labels assigned:")
print(kmeans_labels[:20])
```

### Question 13: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from sklearn.metrics import f1_score
from scipy.stats import mode
import numpy as np

# 1. Map cluster labels to the most frequent true label in that cluster
# We create an array to store the predicted labels that match the original truth
labels_mapped = np.zeros_like(kmeans_labels)

for i in range(10):
    # Find the mask for current cluster
    mask = (kmeans_labels == i)

    # Get the actual labels for the images in this cluster
    true_labels_in_cluster = y[mask]

    if len(true_labels_in_cluster) > 0:
        # Find the most frequent label in this cluster
        most_frequent = mode(true_labels_in_cluster, keepdims=True)[0][0]

        # Map all points in this cluster to this label
        labels_mapped[mask] = most_frequent

# 2. Calculate the macro-averaged F1 score
f1 = f1_score(y, labels_mapped, average='macro')

print(f"Macro-averaged F1 score: {f1:.4f}")
```
